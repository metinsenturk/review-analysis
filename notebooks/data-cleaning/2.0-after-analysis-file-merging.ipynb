```python
import numpy as np
import pandas as pd

complete_path = '../../data/processed/hi_rws_0001_0256_complete.csv'
topics_path = '../../data/processed/hi_rws_0001_0256_topics_selected.csv'
sentiments_path = '../../data/processed/hi_rws_0001_0256_sentiments.csv'

df1 = pd.read_csv(complete_path)
df2 = pd.read_csv(topics_path)
df3 = pd.read_csv(sentiments_path)

df1 = df1.drop(['sent_tokens', 'word_tokens', 'norm_tokens', 'sent_topics'], axis=1)
df1 = df1.drop(['word_tokens_doc', 'norm_tokens_doc'], axis=1)
df1 = df1.drop(['topic_mode', 'topic_list', 'svm_classifier'], axis=1)
df2 = df2.drop(['Unnamed: 0'], axis=1)
df3['sgd.1'].name = 'sentiment_score'

df = pd.concat([df1, df2, df3['sgd.1']], axis=1)
df.info()
```

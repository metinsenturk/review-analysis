```python
import numpy as np
import pandas as pd

file_path = "../../data/raw/yp_competitors_rws.csv"
file_path = "../../data/processed/yp_kimos-maui-lahaina_rws_preprocessed.csv"
df = pd.read_csv(file_path) 
df.shape
```

```python
df.head()
```

```python
from itertools import chain
from nltk import FreqDist
import ast
```

```python
# csv list fix with : ast
df.norm_tokens = df.norm_tokens.apply(lambda x: ast.literal_eval(x))
df.sent_tokens = df.sent_tokens.apply(lambda x: ast.literal_eval(x))
df.pos_tag = df.pos_tag.apply(lambda x: ast.literal_eval(x))
df.word_tokens = df.word_tokens.apply(lambda x: ast.literal_eval(x))
```

```python
df.head()
```

## Create Vocabulary

```python
vocab = sorted(set(chain(*df.norm_tokens)))
len(vocab)
```

```python
# bag of words/ binary
df_features = df.apply(
    lambda x: pd.Series(
        1 if w in x.norm_tokens else 0 
        for w in vocab
    ), axis=1)
df_features.columns = vocab
df_features.head()
```

```python
# bag of words/ frequency
df_features_freq = df.apply(
    lambda x: pd.Series(
        x.norm_tokens.count(w) if w in x.norm_tokens else 0 
        for w in vocab
    ), axis=1)
df_features_freq.columns = vocab
df_features_freq.head()
```
